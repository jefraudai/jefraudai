```python
import pandas as pd
import matplotlib.pyplot as plt

#Base = Path(__file__).parent
cwd = Path.cwd()
print(cwd)

# Charger le fichier
df = pd.read_csv("../data/raw/enedis_agreg_202603311914.csv", sep=";")

# Convertir la date
df["date"] = pd.to_datetime(df["date"])

```

```python
import matplotlib.pyplot as plt

# Agrégation
agg = (
    df.groupby(["date", "region"])["total_energie_soutiree_wh"]
    .sum()
    .reset_index()
)

# Pivot pour tracer les régions
pivot = agg.pivot(index="date", columns="region", values="total_energie_soutiree_wh")
pivot.sort_index().plot(figsize=(16,6))

plt.xlabel("Date")
plt.ylabel("Total énergie soutirée (Wh)")
plt.title("Consommation totale par région")

plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()

plt.show()

# Afficher un graphique par région
# for region in pivot.columns:
#     plt.figure(figsize=(16, 6))
#     pivot[region].plot()
#     plt.xlabel("Date")
#     plt.ylabel("Total énergie soutirée (Wh)")
#     plt.title(f"Consommation totale pour la région {region}")
#     plt.tight_layout()
#     plt.show()

```

```python
df = df[df['region'] == 'Bourgogne-Franche-Comté']

agg = (
    df.groupby(["date", "secteur_activite"])["total_energie_soutiree_wh"]
    .sum()
    .reset_index()
)

pivot = agg.pivot(index="date", columns="secteur_activite", values="total_energie_soutiree_wh")
pivot = pivot[pivot.columns[::-1]]  # Reverse the order of columns
pivot.sort_index().plot(kind='area', stacked=True, figsize=(16,6))

plt.xlabel("Date")
plt.ylabel("Total énergie soutirée (Wh)")
plt.title("Consommation totale par secteur d'activité (Bourgogne)")

plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()

plt.show()
```
